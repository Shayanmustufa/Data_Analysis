As we can see that there are missing values in the data set.To fill the missing values we will take into notice some constraints.

1-Normally Distibiited: IF the data after the visualization is bell shaped and normally distributed,then we will use mean for missing values.
2-Outliers: If there are outliers in the data then we will use median.
3-Skewed: If the data is skewed whether right or left skewed,we will use median in this condition.


For categorical datas ,such as "yes" or "no" or "male" or "female",we will go with the mode.

```python
print(df[df.isnull().any(axis=1)])
print(df[df.isnull().any(axis=1)])
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()
```

Visualizing the numerical columns,to get help in filling out the missing values.

```python
mode_value = df['Gender'].mode()[0]

df['Gender'] = df['Gender'].fillna(mode_value)

mode_value = df['Exercise Habits'].mode()[0]

df['Exercise Habits'] = df['Exercise Habits'].fillna(mode_value)

mode_value = df['Smoking'].mode()[0]

df['Smoking'] = df['Smoking'].fillna(mode_value)

mode_value = df['Gender'].mode()[0]

df['Smoking'] = df['Smoking'].fillna(mode_value)

mode_value = df['Diabetes'].mode()[0]

df['Diabetes'] = df['Diabetes'].fillna(mode_value)

mode_value = df['High Blood Pressure'].mode()[0]

df['High Blood Pressure'] = df['High Blood Pressure'].fillna(mode_value)

mode_value = df['Gender'].mode()[0]

df['Low HDL Cholesterol'] = df['Low HDL Cholesterol'].fillna(mode_value)

mode_value = df['High LDL Cholesterol'].mode()[0]

df['High LDL Cholesterol'] = df['High LDL Cholesterol'].fillna(mode_value)

mode_value = df['Alcohol Consumption'].mode()[0]

df['Alcohol Consumption'] = df['Alcohol Consumption'].fillna(mode_value)

mode_value = df['Stress Level'].mode()[0]

df['Stress Level'] = df['Stress Level'].fillna(mode_value)

mode_value = df['Sugar Consumption'].mode()[0]

df['Sugar Consumption'] = df['Sugar Consumption'].fillna(mode_value)

mode_value = df['Heart Disease Status'].mode()[0]

df['Heart Disease Status'] = df['Heart Disease Status'].fillna(mode_value)
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns


df[numeric_cols] = df[numeric_cols].apply(lambda col: col.fillna(col.mean()))

print(df)
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()
```
